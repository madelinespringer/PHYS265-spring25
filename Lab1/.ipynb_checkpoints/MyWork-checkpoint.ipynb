```python
# Import required Python environments
import matplotlib.pyplot as plt
import numpy as np 
import math

# Create table of constants
G = 6.67*(10**-11) # Gravitational Constant
g = 9.81 # Gravitational Acceleration
Me = 5.9*(10**24) # Mass of Earth
Mm = 7.3*(10**22) # Mass of Moon
Ma = 5500 # Mass of Apollo Command Module
Re = 6378*(10**3) # Radius of Earth
Rm = 1737*(10**3) # Radius of Moon
D = 3.8*(10**8) # Distance from Earth to Moon
Ve = 2.4*(10**3) # Exhaust Velocity of Saturn V Stage 1 (S1-C)
m = 1.3*(10**4) # Burn Rate of S1-C
m0 = 2.8*(10**6) # Wet mass of S-1C
mf = 7.5*(10**5) # Dry mass of S-1C

# Gravitational Potential Function
def grav_potential(M, xm, ym, x, y):
    r = np.sqrt((x-xm)**2 - (y-ym)**2)
    r[r == 0] = np.nan
    return (-G*M)/r

# Plot of Gravitational Potential v. Distance From Earth 
fig, ax = plt.subplots()
x1 = np.linspace(0, 1.5*D, 100)
ax.plot(x1, np.abs(grav_potential(Me, 0, 0, x1, 0)))
ax.set_yscale('log')

ax.set_xlabel('Distance x From Earth (m)')
ax.set_ylabel('Absolute Value of Gravitational Potential (m**2/s**2)')
ax.set_title('Gravitational Potential v. Distance From Earth')
```

```python
# Import required Python environments
from matplotlib.colors import LogNorm
import matplotlib.cm as cm

# 2D Color-Mesh Plot 
def egp(M, xm, ym, x, y):
    r = np.sqrt((x-xm)**2 + (y-ym)**2)
    r[r == 0] = np.nan
    return (-G*M)/r

def mgp(M, xm, ym, x, y):
    r = np.sqrt((x-xm)**2 + (y-ym)**2)
    r[r == 0] = np.nan
    return (-G*M)/r

mx = D/np.sqrt(2) 
my = D/np.sqrt(2)

fig, ax1 = plt.subplots() 
X = np.linspace(-1.5*D, 1.5*D, 100) 
Y = np.linspace(-1.5*D, 1.5*D, 100) 
x_grid, y_grid = np.meshgrid(X, Y) 

Earth = np.abs(egp(Me, 0, 0, x_grid, y_grid))
Moon = np.abs(egp(Mm, mx, my, x_grid, y_grid))
phi = Earth + Moon

colormesh = ax1.pcolormesh(x_grid, y_grid, phi, norm = LogNorm())
ax.set_aspect('equal') 
cbar = fig.colorbar(colormesh) 

ax.set_title("Gravitational Potential of the Earth and Moon System")
ax.set_xlabel("x (m)") 
ax.set_ylabel("y (m)") 
cbar.set_label("Gravitational Potential (J/kg)")
```

```python
# 2D Contour Plot 
fig, ax2 = plt.subplots() 
levels = np.logspace(np.log10(np.min(phi[phi > 0])), np.log10(np.max(phi)), num = 10)
contour = ax2.contour(x_grid, y_grid, phi, levels = levels, norm = LogNorm())
ax2.set_aspect("equal")
cbar = fig.colorbar(contour)

ax2.set_title("Gravitational Potential of the Earth and Moon System") 
ax2.set_xlabel("x (m)") 
ax2.set_ylabel("y (m)")
cbar.set_label("Gravitational Potential (J/kg)")
```

```python
## Gravitational Force Function
def grav_force(M1, m2, x1, y1, x2, y2):
    dx = x2 - x1 
    dy = y2 - y1
    r = np.sqrt(dx**2 + dy**2)

    if r == 0: 
        raise ValueError("Singularity error: Objects cannot be at the same position.")

    F_mag = G*M1*m2/r**2
    Fx = F_mag * (dx/r) 
    Fy = F_mag * (dy/r) 

    return Fx, Fy

## Streamplot of Gravitational Force 
x_range = np.linspace(-1.5*D, 1.5*D, 30)
y_range = np.linspace(-1.5*D, 1.5*D, 30)
x, y = np.meshgrid(x_range, y_range) 

Fx, Fy = np.zeros_like(x), np.zeros_like(y)

for i in range(x.shape[0]): 
    for j in range(x.shape[1]):
        fx_earth, fy_earth = grav_force(Me, 1, 0, 0, x[i, j], y[i, j])
        fx_moon, fy_moon = grav_force(Mm, 1, D, 0, x[i, j], y[i, j])

mag = np.sqrt(Fx**2 + Fy**2) 
Fx /= mag 
Fy /= mag 

fig, ax3 = plt.subplots()
strm = ax.streamplot(x, y, Fx, Fy, color = mag, linewidth = 1, cmap = "plasma")
cbar = plt.colorbar(strm.lines) 
cbar.set_label("Force Magnitude (N/kg)")

ax3.scatter(0, 0, color = 'blue', s = 100, label = "Earth") 
ax3.scatter(D, 0, color = 'gray', s = 50, label = "Moon") 
ax3.set_xlabel("x (m)") 
ax3.set_ylabel("y (m)") 
ax3.set_title("Gravitational Force Field of the Earth and Moon") 
ax3.legend() 
```

```python
## Calculate Burn Time
def burn_time(m0, mf, mdot):
    return (m0 - mf) / mdot

## Velocity Change Formula
def vel_change(t, m0, mf, mdot, Ve, g):
    m_t = m0 - mdot * t
    if m_t < mf: 
        return 0
    return Ve * np.log(m0/m_t) - g * t

## Rocket's Altitude After Burn 
def altitude(m0, mf, mdot, Ve, g):
    T = burn_time(m0, mf, mdot) 
    h, _ = quad(vel_change, 0, T, args = (m0, mf, mdot, Ve, g))
    return h
```
